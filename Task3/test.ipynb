```python
def factorial(n):
    fact = 1
    for num in range(2, n + 1):
        fact *= num
    return fact

factorial(4)
```

```python
def fibo(N, M):
    fib = [0, 1]
    result = []
    
    # Keep generating Fibonacci numbers while the latest is less than N
    while fib[-1] < N:
        if fib[-1] % M == 0:
            result.append(fib[-1])
        fib.append(fib[-1] + fib[-2])
    
    # Check the first number (0) as well
    if 0 < N and 0 % M == 0:
        result.insert(0, 0)
    
    return result

print(fibo(100, 7))
```
