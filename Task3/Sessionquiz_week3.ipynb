```python
from math import sqrt
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]

radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]

names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

# Calculating escapevelocity for each object
number_of_objects = len(masses)
range_list = range(number_of_objects)
escape_list = []
for i in range_list:
    v = sqrt((2*G*(masses[i]))/(radii[i]))
    print(f"The escape velocity of {names[i]} with mass {masses[i]:6.2e} is: {v:6.2e}")
    escape_list.append(
        v)  # Puts each v into a list

print() 

# Prints all the names and equivalent values in to one dictionary
escape_dictionary = dict(zip(names, escape_list)) 
print(escape_dictionary)


```

```python
from math import atan, degrees

# Defines our function delta
def ang_size(d, D):
    """The defined function calculates the angular size 'delta' of an object
       The input values are the size of the object 'd' and the distance to the object 'D'
    """
    atn = atan(d/(2*D))
    delta = 2 * degrees(atn)
    return delta
    
print(f"The angular size for the given values is: {ang_size(7, 6)} degrees.")
```

### 3.10 (to be checked)

```python
from scipy.constants import pi

def cylinder(r, h):
    """The code calculates the area 'A' and the volume 'V' of a cylinder
       The input values are the radius 'r' and the height 'h'
    """
    A = 2 * pi * (r**2) + 2* pi * r * h
    V = pi * (r**2) * h
    return A, V

area, volume = cylinder(10, 3)
print(f"The area of the cylinder is: {area} and it's volume is: {volume}")

# Calculating the ratio between the area and volume
ratio = area/volume
print(f"The ratio between the area and volume is: {ratio}")
```

### 3.17 (to be checked)

```python
import numpy as np

# Creating definition to calculate the standard deviation
o = [3, 2, 7, 3, 8, 9, 10, 5, 2]
b = np.array(o)

def stdev(b):
    """Calculates the standard deviation using NumPy
    """
    a = np.mean(b)  # Calculates mean
    N = np.prod(b.shape)  # Gives product of the arrays dimensions
    s = np.sum((b - a)**2) # Summing 
    std = np.sqrt(s/N)
    return std

print(f"The standard deviation calculated using the definition is: {stdev(b)}")

# Standard deviation using the NumPy function
print(f"The standard deviation using NumPy function is: {np.std(b)}")

```

```python
# One liner
o = [3, 2, 7, 3, 8, 9, 10, 5, 2]
b = np.array(o)

def stdev(b):
    """Calculates the standard deviation using NumPy
    """
    std = np.sqrt(np.sum((b - np.mean(b))**2)/(np.prod(b.shape)))
    return std

print(f"The standard deviation calculated using the one liner definition is: {stdev(b)}")
```

### 3.22 (to be checked)

```python
from matplotlib import pyplot as plt
import numpy as np

# Plotting catenary
y = []
r = range(-5, 6)
x = [i for i in r]

for i in r:
    a = np.cosh(i)
    y.append(a)

plt.plot(x, y, color = 'red', label = r"$\cosh{x}$")
plt.grid()
plt.title("Catenary")
plt.ylabel("Weight", fontsize = 13)
plt.xlabel("Location", fontsize = 13)
plt.legend(fontsize = 13)
plt.show()

    
```
