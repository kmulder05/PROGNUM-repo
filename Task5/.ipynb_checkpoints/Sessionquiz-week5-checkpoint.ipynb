### 5.2 (to be checked)

```python
import numpy as np

# Function for getting the arrays
def load_arrays(filename):
    """Reads the file and sorts the data in to two arrays"""
    data = np.loadtxt(filename, comments = '#')  # Loads text file, comments excludes anything starting with the given input
    x = data[:, 0]
    y = data[: , 1]
    return x, y

# User can input filename
filename = input(f"Enter the name of your file:")  # Filename: excercise.txt

# Printing both arrays
x, y = load_arrays(filename)
print(f"The values for x are: {x}, with mean: {np.mean(x)}, minimum value: {np.min(x)} and maximum: {np.max(x)}")
print(f"The values for y are: {y}, with mean: {np.mean(y)}, minimum value: {np.min(y)} and maximum: {np.max(y)}")

# The new array: y minus median
y2 = (y - np.median(y))**2
print(f"New array y2: {y2}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def load_arrays(filename):
    """Reads the file and sorts the data in to two arrays"""
    data = np.loadtxt(filename, comments = '#')  # Loads text file, comments excludes anything starting with the given input
    x = np.array(data)
    return x
    
filename = "ngc6946.txt"  
x = load_arrays(filename)
x.shape = (40, 50, 70)  # Shapes our data
slc = x[:,25,35]  # Slices the line along z at a certain x and y value
print(f"The z-profiel in the middle is: {slc}")

# Elementary statistics for the profile
print(f"The maximum is: {np.max(slc)}")
print(f"The minimum is: {np.min(slc)}")
print(f"The sum is: {np.sum(slc)}")
print(f"The mean is: {np.mean(slc)}")
print(f"The standard deviation is: {np.std(slc)}")

# Plotting sliced profile
plt.plot(slc)
plt.text(11, np.max(slc), 'Maximum')
plt.show()
```

### 5.11 (to be checked)

```python
import numpy as np
import matplotlib.pyplot as plt

def load_arrays(filename):
    """Reads the file and sorts the data in to two arrays"""
    data = np.loadtxt(filename, comments = '#')  # Loads text file, comments excludes anything starting with the given input
    x = np.array(data)
    return x
    
filename = "ngc6946.txt"  
x = load_arrays(filename)
x.shape = (40, 50, 70)  # Shapes our data
slc = x[0,:,:]

# Mean, mean + std and mean + 2 * std
a = np.mean(slc)
b = np.std(slc)
c = a + b
d = a + 2 * b


plt.imshow(slc, origin = 'lower')
plt.colorbar()
plt.contour(slc, levels = [a, c, d])
```

### 5.12 (to be checked)

```python
from scipy.constants import c

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    lightspeed = c
    
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    def velocity(self):
        z = self.redshift()  # Calls calculated redshift from above
        v = z * self.lightspeed  # Calls the constant lightspeed
        v_km_s = v/1000  # Converts m/s to km/s
        return f"The calculated speed of the object is: {v_km_s} km/s"
        
gal1 = Galaxy("NGC1832", 3994.7)
print(gal1.velocity())
```

### 5.13 (to be checked)

```python
import numpy as np
from matplotlib.pyplot import figure, show
from matplotlib.patches import Ellipse

# Creating dummy array
Nx = 400
Ny = 300
data = np.random.random((Ny,Nx))

# Conversion of size of ellips with AstroEllipse 
class AstroEllipse(Ellipse):
   def __init__(self, xc, yc, semi_major, semi_minor, position_angle, **args):
      assert(semi_major >= semi_minor)  # Prerequisite to make angle unambiguous
      XY = (xc, yc)                  # MPL requires center as a tuple
      angle_mpl = position_angle     # Angle for MatPlotLib's ellipse patch
      height = 2.0*semi_major        # height: is along vertical axis for an unrotated ellipse
      width  = 2.0*semi_minor        # width:  is along horizontal axis for an unrotated ellipse
      super().__init__(XY, width, height, angle=angle_mpl, **args)

fig = figure()
frame = fig.add_subplot(1,1,1, aspect=1)
xcenter = Nx/2
ycenter = Ny/2
semimajor = 100;
semiminor= 50
theta = 30    # degrees 
ell = AstroEllipse(xcenter, ycenter, semimajor, semiminor, theta, fill=False, hatch='/')
frame.add_patch(ell)
frame.set_xlim(0,400)
frame.set_ylim(0,300)
frame.grid()
plt.imshow(data, interpolation='none', origin="lower", cmap='gray')
show()
```
