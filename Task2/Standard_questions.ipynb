```python
from math import *

# Given values
x = 2
y = 3

# Calculating angle between vector and x-axis
phi = atan2(y, x)
dgr = degrees(phi)
print(f"y over x in radians is: {phi}")
print(f"The angle in degrees is: {dgr}")

# Calculating the answer of y/x and seeing if that equals the tangent of the angle.
tan_y_x = tan(phi)
print(f"The tangent of the calculated angle gives : {tan_y_x}, which equals y/x, which is: {y/x}")

```
