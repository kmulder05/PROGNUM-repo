### 2.12 (checked)

```python
from math import *

# Calculating the angle between the line and the x-axis
x = float(input(f"Give a value for x:"))
y = float(input(f"Give a value for y:"))
ang = degrees(atan2(y, x))

# Asigning the coordinates to the right quadrant.
if x > 0 and y > 0:
    print(f"Your coordinates are in the first quadrant and the angle between the line and the x-axis is: {ang} degrees.")
elif x < 0 and y > 0:
    print(f"Your coordinates are in the second quadrant and the angle between the line and the x-axis is: {ang} degrees.")
elif x < 0 and y < 0:
    print(f"Your coordinates are in the third quadrant and the angle between the line and the x-axis is: {180 + ang} degrees.")
else:
    print(f"Your coordinates are in the fourth quadrant and the angle between the line and the x-axis is: {180 + ang} degrees.")
    
```

What is special about atan2() compared with math.atan(): atan2() can calculate the angle in any quadrant while math.atan() can only calculate the angle in the first quadrant. Atan2() also takes two arguments while math.atan() only takes 1.

### 2.14 (checked)

```python

# Creating a list
x = [1.36,  1.46,  1.41,  2.12,  0.42,  0.98,  0.45,  1.97, -0.24, 2.18]
print(x)

# Add 0.51 to the list and remove 2.12
x.append(0.51)
x.remove(2.12)
print(x)

# Find value at index 6
value_i6 = x[6]
print(f"The value at index 6 is: {value_i6}")

# Find index of value 1.46
a = x.index(1.46)
print(f"The index of 1.46 is: {a}")

# Using slice from n=2, including all even indexes except last even index
indices = slice(0, 7, 2)
x_slice = x[indices]
print(x_slice)

# Calculating sum of slice
s = sum(x_slice)
print(s)
```

### 2.15 (checked)

```python

# Converting lists to sets
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
set_night1 = set(night1)
set_night2 = set(night2)
print(set_night1)
print(set_night2)

# Finding common values in both sets
inters = set_night1.intersection(set_night2)
print(f"Common values between both sets night1 and night 2: {inters}")

# Rewriting the set "inters" as a list.
nights_list = list(inters)
print(nights_list)

# Using the in operator to find whether 5 is in the common values list
a = 5 in nights_list
print(f"Is there a 5 in the list of common values?: {a}")

# Using in operator to find whether 5 is in the set version of night1
b = 5 in set_night1
print(f"Is there a 5 in the set of night 1?: {b}")
```
