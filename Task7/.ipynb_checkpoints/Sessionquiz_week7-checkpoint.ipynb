```python
import numpy as np
from numpy import linalg as la

# Creating and shaping A
A = np.array([1, 2, 3, 0, -4,  1, 0, 3, -1])
A.shape = (3,3)
print("This is matrix A:")
print(A)

print()

# Finding the determinant of A
det = la.det(A)
print(f"The determinant of A is: {det}")

print()

# Finding solutions to given equations
b = np.array([2, 2, -1])  # Creating vector b
A_inv = la.inv(A)  # Finding inverse
xyz = A_inv@b  # Augmentend matrix to find solutions
print(f"The values for x, y, z are: {xyz}")

print()

# Finding whether A is orthogonal
A_transp = A.T
print(f"Is A orthogonal: {np.allclose(A_transp, A_inv)}")

print()

# Finding the eigenvalues
Lambda, v = la.eig(A)
print(f"The Eigenvalues are: {Lambda}")
```

```python
from numpy import linalg as la
import numpy as np
from math import *

# Converting angles to radians
angle_1 = np.radians(180-123)
angle_2 = np.radians(90-27.4)
angle_3 = np.radians(192.25)

# All E components
Ex = np.array([1,0,0,0,cos(angle_1),sin(angle_1),0,-sin(angle_1),cos(angle_1)]).reshape(3,3)
Ey = np.array([cos(angle_2),0,-sin(angle_2),0,1,0,sin(angle_2),0,cos(angle_2)]).reshape(3,3)
Ez = np.array([cos(angle_3),sin(angle_3),0,-sin(angle_3),cos(angle_3),0,0,0,1]).reshape(3,3)

Ez_1 = np.array([cos(angle_1),sin(angle_1),0,-sin(angle_1),cos(angle_1),0,0,0,1]).reshape(3,3)

E = Ez_1 @ Ey @ Ez
print(E)

print()

# Checking for orthogonality
E_inv = la.inv(E)
E_transp = E.T
print(f"Is this vector orthogonal?: {np.allclose(E_transp, E_inv)} and is determinant should be 1: {la.det(E)}")

print()

# Finding Eigenvalues
Lambda, v = la.eig(E)
print(Lambda)  # There is an eigenvalue equal to 1

print()

# Finding v
alpha = np.radians(192.25)
delta = np.radians(27.4)
v1 = np.array([cos(delta)*cos(alpha), cos(delta)*sin(alpha), sin(delta)]).reshape(3,1)
print("v1 = ")
print(v1)
print(f"The norm of v1 = {la.norm(v1)}")

print()

# Calculating v2
v2 = E @ v1
print(np.round(v2))
```

### 7.5 (unchecked)

```python
import numpy as np
from numpy.polynomial.polynomial import polyvander
from matplotlib.pyplot import *

x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])

# Calculating Vandermonde matrix and finding coefficients
vdm = polyvander(x,(len(x)-1))
a = np.linalg.solve(vdm, y)
print(f"The polynomials are: {a}")

# Preparation for the the coefficients plot
x_plot = np.linspace(x[0], x[-1], 500)  # Creates a plot of 500 values between upperbound and lower bound of given array, for a smoother curve
y_plot = np.zeros_like(x_plot)  # Creating as much zeros as there are values in x_plot
for i in range(len(x)):  # Calculating the polynomials for bounds within the plot
    y_plot += a[i] * x_plot**i

# Plotting the figure using OOP
fig = figure()
frame = fig.add_subplot(1,1,1)
frame.scatter(x, y, c ='r')
frame.plot(x_plot, y_plot)
frame.set_title("Polynomials plot")
frame.set_xlabel("X")
frame.set_ylabel("Y")
show()
```

```python
import numpy as np
from scipy import interpolate as ip
from matplotlib.pyplot import figure, show
from numpy.polynomial.polynomial import polyvander

x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])

# Cubic spline
cs = ip.interp1d(x, y, kind = 'cubic')

newx = np.linspace(x[0], x[-1], 250)
newy = cs(newx)

# Plotting the figure
fig = figure()
frame = fig.add_subplot(1,1,1)
frame.plot(x, y, 'go')
frame.plot(newx, newy, 'r')
frame.grid(True)
frame.set_xlabel("X")
frame.set_ylabel("Y")
frame.set_title("Cubic Split")
show()

# This way is much easier, shorter and faster than using the polynomial interpolation

class interpolation:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def linear_int(self, x_hat):  # Calculating the linear interpolation 
        """Linear interpolation"""
        li = ip.interp1d(self.x, self.y, kind = 'linear')
        return li(x_hat)

    def polynomial_int(self, x_hat):  # Calculates the polynomial interpolation
        """Polynomial interpolation"""
        coeffs = np.polyfit(self.x, self.y, len(self.x) - 1)
        pi = np.poly1d(coeffs)
        return np.polyval(pi, x_hat)

    def spline_int(self, x_hat):  # Calculates the spline interpolation
        """Spline interpolation"""
        si = ip.interp1d(self.x, self.y, kind = 'cubic')
        return si(x_hat)

    def find_y_hat(self, x_hat, method = 'spline'):  # Calculates the value of y_hat for a given x_hat and uses one of the functions in the class to calculate it
        """Finding y_hat for a given x_hat"""
        if method == 'linear':
            return self.linear_int(x_hat)
        elif method == 'poly':
            return self.polynomial_int(x_hat)
        elif method == 'spline':
            return self.spline_int(x_hat)
        else:
            return "Not the right method" 

setup1 = interpolation(x, y)
print(f"The value for y_hat is: {setup1.find_y_hat(2.5, 'linear')}")
print(f"The value for y_hat is: {setup1.find_y_hat(2.5, 'poly')}")
print(f"The value for y_hat is: {setup1.find_y_hat(2.5, 'spline')}")
```

### 7.10 (unchecked)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq, curve_fit

# 1. Data Setup
data = [
   18.49,     2094.75,    0.3,           100,
   35.48,     2320.87,    2.2,           120,
   35.48,     1224.9,     3.0,           80,
   11.34,     872.13,     1.2,           80,
   7.12,      432.2,      0.4,           70,
   9.82,      751,        0.3,           75,
   23.66,     1410,       2.3,           70,
   44,        2700,       4.0,           110,
   66.7,      5292,       5.3,           120,
   52.4,      2550,       5.2,           100,
   55,        5253,       6.1,           150,
   25.6,      2419,       3.2,           110,
   26.19,     2158,       2.8,           110
]
data_list = np.array(data)
d = data_list[0::4]
v = data_list[1::4]
err_v = data_list[3::4]

# 2. Define Models
# For leastsq: beta is a list [intercept, slope]
def model_lsq(beta, d):
    return beta[0] + beta[1] * d

def residuals(beta, d, v, err_v):
    return (v - model_lsq(beta, d)) / err_v

# For curve_fit: parameters are individual
def model_cf(d, b0, b1):
    return b0 + b1 * d

# 3. Perform Fits
beta_start = [0, 70.0]

# Fit 1: leastsq (Weighted)
res_lsq, _ = leastsq(residuals, beta_start, args=(d, v, err_v))

# Fit 2: curve_fit (Weighted)
popt, pcov = curve_fit(model_cf, d, v, p0=beta_start, sigma=err_v, absolute_sigma=True)

# 4. Plotting
plt.figure(figsize=(10, 6))

# Plot data points with error bars and a different marker
plt.errorbar(d, v, yerr=err_v, fmt='s', color='black', label='Observed Data', capsize=3)

# Generate line for plotting
x_fit = np.linspace(0, d.max() + 5, 100)

# Plot Leastsq result
label_lsq = f"leastsq: v = {res_lsq[0]:.2f} + {res_lsq[1]:.2f}d"
plt.plot(x_fit, model_lsq(res_lsq, x_fit), 'r--', label=label_lsq)

# Plot Curve_fit result
label_cf = f"curve_fit: v = {popt[0]:.2f} + {popt[1]:.2f}d"
plt.plot(x_fit, model_cf(x_fit, *popt), 'g:', linewidth=2, label=label_cf)

plt.xlabel('Distance (Mpc)')
plt.ylabel('Velocity (km/s)')
plt.title("Hubble's Law Best-Fitted Parameters")
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, None)
plt.ylim(0, None)

plt.show()

print(f"Leastsq Params: Intercept={res_lsq[0]:.2f}, Slope={res_lsq[1]:.2f}")
print(f"Curve_fit Params: Intercept={popt[0]:.2f}, Slope={popt[1]:.2f}")
```
